```python
print('常態分配')
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as nr
x=nr.normal(5,1,100) #平均值為5，標準差為1 #normal(平均值,標準差,數量)
y=nr.normal(5,2,100) #平均值為5，標準差為2
z=nr.normal(5,3,100) #平均值為5，標準差為3
plt.hist(z,color='b')
plt.hist(y,color='g')
plt.hist(x,color='r')
plt.show()
```

```python
import numpy.random
print(numpy.random.rand())
print()
#0到1之間隨機產生1個數值
print(numpy.random.rand(10))
print()
#0到1之間隨機產生10個數值
print(numpy.random.rand(10,10))
print()
#0到1之間隨機產生10*10的矩陣
print(numpy.random.rand(100)*40+30)
#產生100個 0到1之間的數值，*40 代表0到40之間
# +30 代表 30到70之間的數值
```

```python
import numpy as np
import matplotlib.pyplot as plt
print(np.max(money))
print(np.min(money))
print('110個職業類別，他們薪資的分布可否用圖表呈現')
plt.hist(money) #hist直方圖
plt.vlines(np.mean(money),0,40,color='red',label='mean') #vlines直線
plt.vlines(np.median(money),0,40,color='yellow',label='median')
plt.legend()
plt.show()
```
