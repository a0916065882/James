```python
import numpy as np
import pandas as pd
df = pd.DataFrame(np.arange(16).reshape(4,4),
                   index=['台北','台中', '台南', '高雄'],
                   columns=['one', 'two', 'three', 'four'])
df
```

```python
#drop是用來刪除行或列的方法
df.drop(index=['台北', '高雄'], inplace=True) #inplace是指在原始物件上進行修改
df
```

```python
#排序和排名

scores = np.random.randint(50,101,size=(50, 5))
df = pd.DataFrame(scores,
                  columns=['國文', '英文', '數學', '自然', '社會'],
                  index=['學生' + str(i) for i in range(1, 51)])
df.index.name = '學生' #設定索引名稱
df.columns.name = '科目' #設定列名稱
sum_values = df.sum(axis=1) #加總columns
mean_values = df.mean(axis=1) #平均columns
df['總分'] = sum_values #加入總分列
df['平均'] = mean_values #加入平均列
df
```

```python
#排序和排名
#rank()
rank_values = sum_values.rank(method='min', ascending=False )
df['排名'] = rank_values
df
```

```python
#sort_values()是用來根據值排序的函數
#sort_index()是用來根據索引排序的函數
df1 = df.sort_values(by='總分', ascending=False)
df1
```
