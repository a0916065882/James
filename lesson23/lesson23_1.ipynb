```python
import numpy as np
import matplotlib.pyplot as plt
print('準備產生母體資料，透過程式產生一萬筆資料')
dice=[1,2,3,4,5,6] 
population=[]
#population=np.random.choice(dice,10000) #如此一來沒有連續資料的特性
for x in range(10000): #range(10000)代表產生一萬筆母體資料
  original=np.random.choice(dice,100) #隨機產生100個資料
  population.append(original.mean()) #這一百筆資料取平均值加入到母體
print('查看母體資料')  
print(population)
print('母體平均數:',np.mean(population))
print('母體標準差:',np.std(population))
p1=np.mean(population)
p2=np.std(population)
x=np.random.normal(p1,p2,10000) #normal(平均數,標準差,產生資料筆數)
plt.hist(x,color='red') #依據母體資料繪製的常態分配資料
plt.hist(population,color='blue') #隨機產生的母體資料
plt.show()
```

```python
print('樣本的建立與計算')
print('母體的平均數:',p1)
print('母體的標準差:',p2)
plt.hist(population,color='blue')
sample1=np.random.choice(a=population,size=10) #樣本10個
sample2=np.random.choice(a=population,size=100) #樣本100個
sample3=np.random.choice(a=population,size=1000) #樣本1000個
sample4=np.random.choice(a=population,size=5000) #樣本5000個
plt.hist(sample4,color='m')
plt.hist(sample3,color='y')
plt.hist(sample2,color='g')
plt.hist(sample1,color='r')
plt.show()
print('樣本10的平均數:',np.mean(sample1))
print('樣本100的平均數:',np.mean(sample2))
print('樣本1000的平均數:',np.mean(sample3))
print('樣本5000的平均數:',np.mean(sample4))
```
