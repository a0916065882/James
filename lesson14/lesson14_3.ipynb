```python
py_list = [1, 2, 3, 4]
#一般操作法
new_list = []
for ele in py_list:
    print(ele)
    new_list.append(ele*10)
new_list
```

```python
py_list = [1, 2, 3, 4]
#使用綜合表達式(comprehention)
[ele*10 for ele in py_list]
```

```python
#利用numpy和matplotlib產生正弦波和餘弦波
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='正弦波')
plt.plot(x, y2, label='餘弦波')
plt.legend()
plt.show()
```
