```python
print('查找遺失值1')
print(kk.isnull().sum())#計算遺失值數量
print('查看每個欄位的資料型態')
print('主要目的查看那些欄位是object，object代表文字型態')
print(kk.dtypes) #全部欄位的資料型態
```

```python
print('遺失值乍看之下沒有，我們再來做一個動作確認')
print('分成數值與字串兩種欄位資料紀錄')
list3=[] #數值欄位
list4=[] #字串欄位
for i in kk.columns:
  print('欄位名稱:',i)
  print('欄位資料型態:',kk[i].dtype)
  print('假設資料型態是object，我們得嘗試轉換為float，可以轉換就代表有問題，有遺失值存在')
  if kk[i].dtype==object:
    try:
      kk[i]=kk[i].astype('float') #轉換
      list3.append(i) #轉換成功
      print('轉換成功')
    except:
      print('不能轉換')
      list4.append(i) #文字資料
  else:
    list3.append(i)

print('數值欄位:',list3)
print('文字欄位:',list4)
```

```python
print('欄位名稱與內容看起來沒有資料唯一性的問題，不像customer_id是唯一編號，不進行重複值處理')
print('字串部分，以Bar顯示，分類方式操作')
import matplotlib.pyplot as plt
import seaborn as sns
for cols in list4:
  plt.figure(figsize=(20,16))
  plt.title(cols)
  sns.countplot(x=cols,data=kk,hue='VAX3')
  plt.savefig('bar2_'+cols+'.png')
  plt.show()
```
