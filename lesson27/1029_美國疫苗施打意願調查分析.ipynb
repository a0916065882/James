```python
import pandas as pd
kk=pd.read_csv('KK_Vaccines_replication_data.csv')
print('連結完成')
print(kk.head(10))
print('改為網頁輸出，較方便查看內容')
kk.to_html('kk1.html')
print('了解欄位名稱，-容易有誤判風險')
print(kk.columns)
print('儲存欄位名稱')
cols=kk.columns
kk.columns=[str.replace('-','_') for str in cols]
print('修改完成')
print(kk.columns)
print('改為網頁輸出，較方便查看內容')
kk.to_html('kk2.html')
```

```python
print('查找遺失值1')
print(kk.isnull().sum())#計算遺失值數量
print('查看每個欄位的資料型態')
print('主要目的查看那些欄位是object，object代表文字型態')
print(kk.dtypes) #全部欄位的資料型態
```

```python
print('遺失值乍看之下沒有，我們再來做一個動作確認')
print('分成數值與字串兩種欄位資料紀錄')
list3=[] #數值欄位
list4=[] #字串欄位
for i in kk.columns:
  print('欄位名稱:',i)
  print('欄位資料型態:',kk[i].dtype)
  print('假設資料型態是object，我們得嘗試轉換為float，可以轉換就代表有問題，有遺失值存在')
  if kk[i].dtype==object:
    try:
      kk[i]=kk[i].astype('float') #轉換
      list3.append(i) #轉換成功
      print('轉換成功')
    except:
      print('不能轉換')
      list4.append(i) #文字資料
  else:
    list3.append(i)

print('數值欄位:',list3)
print('文字欄位:',list4)
```

```python
print('欄位名稱與內容看起來沒有資料唯一性的問題，不像customer_id是唯一編號，不進行重複值處理')
print('字串部分，以Bar顯示，分類方式操作')
import matplotlib.pyplot as plt
import seaborn as sns
for cols in list4:
  plt.figure(figsize=(20,16))
  plt.title(cols)
  sns.countplot(x=cols,data=kk,hue='VAX3')
  plt.savefig('bar2_'+cols+'.png')
  plt.show()
```

```python
import pandas as pd
print('1、X進行欄位的準備')
list5=list4.copy()
print(list5)
list5.pop() #刪除list最後一個資料
print(list5)#沒有y的資料
print('2、VAX3得做轉換，y這部分得做轉換')
vax3_mapping={'Yes':1,'No':2}
kk['ans']=kk['VAX3'].map(vax3_mapping)
print(kk['ans'].value_counts())
print(kk['VAX3'].value_counts())
print(kk.dtypes)
print(kk.shape)
print('3、X欄位都是文字型態，所以得做編碼處理')
print('產生一個DataFrame')
kk2=pd.DataFrame()
kk2[list5]=kk[list5]
print(kk2.head())
print('進行編碼處理')
kk3=pd.get_dummies(kk2)
print(kk3.head())
kk3['ans']=kk['ans']
print(kk3.head())
print(kk3.shape)
kk3.to_html('kk3.html')
```
