```python
import yfinance as yf
#台積電(2330.TW),聯電(2303.TW),聯發科(2454.TW),鴻海(2317.TW)
#取得所有 2024 年（1/1 起至 2025/1/1 不含）的交易日資料
tickers = ["2330.TW", "2303.TW", "2454.TW", "2317.TW"]
data = {}
for ticker in tickers:
    stock = yf.Ticker(ticker)
    data[ticker] = stock.history(start="2024-01-01", end="2025-01-01")
data["2330.TW"]["Close"] # 取得台積電收盤價
```

```python
#data有四檔股票的2024年交易日資料，我需要每一個收盤價
import pandas as pd
close_prices = {ticker: df["Close"] for ticker, df in data.items()} #詳細說明程式碼:這行程式碼的作用是從data字典中提取每檔股票的收盤價，並將其存儲在close_prices字典中
#產生DataFrame
close_prices_df = pd.DataFrame(close_prices)
close_prices_df.columns = ["台積電", "聯電", "聯發科", "鴻海"]
close_prices_df
```

```python
#已經有中文字型`ChineseFont.ttf`
#接下來要設定字型
#matplotlib預設字型為ChineseFont.ttf
from matplotlib.font_manager import fontManager
import matplotlib as mpl
fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family='ChineseFont')
```

```python
#close_prices_df,建立線圖
#使用plt畫圖(111),沒有副表格時使用,可快速建立圖表
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
for column in close_prices_df.columns:
    plt.plot(close_prices_df.index, close_prices_df[column], label=column)
plt.title("各股票收盤價")
plt.xlabel("日期")
plt.ylabel("收盤價")
plt.legend()
plt.show()
```

```python
#標準寫法
#有多個圖表時,需用ax建立
figure = plt.figure(figsize=(12, 6))
ax = figure.add_subplot(111)
for column in close_prices_df.columns:
    ax.plot(close_prices_df.index, close_prices_df[column], label=column)
ax.set_title('收盤價')
ax.set_xlabel('日期')
ax.set_ylabel('價格')
ax.legend()
plt.show()
```

```python
#close_prices_df,要顯示每日的漲幅金額,例如漲3元,跌-3元
daily_changes = close_prices_df.diff() # diff() 計算每日變化(原資料往下偏移一格後相減)
daily_changes
```

```python
#close_prices_df,要顯示每日的漲幅比例,例如漲3%,跌-3%
daily_changes = close_prices_df.pct_change() * 100 # pct_change() 計算每日變化百分比
#顯示要有%符號
daily_changes = daily_changes.applymap(lambda x: f"{x:.2f}%" if pd.notnull(x) else "-")
daily_changes
```
